# 2 RWs Collab on ER

```python
collab_covered_avg1 = np.loadtxt(folder_path+"2rw_vs_1_rw_on_er_collab.csv", delimiter=',')
single_covered_avg1 = np.loadtxt(folder_path+"2rw_vs_1_rw_on_er_single.csv", delimiter=',')
```

```python
c = 5
cs = np.arange(c, step=0.01)

diff1 = collab_covered_avg1 - np.array([single_covered_avg1[2*i] for i in range(c*n+1)])
scaled_diff1 = np.array([diff1[int(i*n)] for i in cs])
plt.plot(cs, scaled_diff1, label='Difference: Collab - Single')
plt.title("Visited Vertices Difference vs. c")
plt.xlabel("c")
plt.ylabel("Visited Vertices")
plt.legend(title=f"n={n}, k={k}, ER")
plt.show()
```

## Testing 1/c

```python
def f1(c, a1, a2, a3):
    return a1/(c + a2) + a3
```

```python
popt, pcov = curve_fit(f1, cs, scaled_diff1)

fitted = f(cs, *popt)
```

```python
plt.plot(cs, scaled_diff1, label='Difference: Collab - Single')
plt.plot(cs, fitted, label='Fitted Curve')
plt.title("Visited Vertices Difference vs. c")
plt.xlabel("c")
plt.ylabel("Visited Vertices")
plt.legend(title=f"n={n}, k={k}, ER")
plt.show()
```

```python
def MSE(A, B):
    n = len(A)
    return ((A - B)*(A - B)).mean()
```

```python
print(f"MSE: {MSE(scaled_diff1, fitted)}")
```

## Testing 1/c^2

```python
def f2(c, a1, a2, a3):
    return a1/(c*c + a2) + a3
```

```python
popt, pcov = curve_fit(f2, cs, scaled_diff1)

fitted = f2(cs, *popt)
```

```python
plt.plot(cs, scaled_diff1, label='Difference: Collab - Single')
plt.plot(cs, fitted, label='Fitted Curve')
plt.title("Visited Vertices Difference vs. c")
plt.xlabel("c")
plt.ylabel("Visited Vertices")
plt.legend(title=f"n={n}, k={k}, ER")
plt.show()
```

```python
print(f"MSE: {MSE(scaled_diff1, fitted)}")
```

## Testing e^(-c)

```python
def f3(c, a1, a2, a3):
    return a1*np.exp(-a2*c) + a3
```

```python
popt, pcov = curve_fit(f3, cs, scaled_diff1)
fitted = f3(cs, *popt)
```
