# Collaboration

This notebook explores collaboration between two walkers on graphs.

```python
# Setup

%matplotlib inline
import sys
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import random

sys.path.insert(0, '..')
plt.rcParams['figure.figsize'] = [10, 8]

import walkers
from graph import Graph
from simulator import Simulator
```

```python
resimulate = False
```

## Collaboration with 2 Random Walks on an ER Random Graph

```python
# Parameters
n = 1000
d = 10
k = 1000
steps = 5*n
```

```python
if resimulate:
    collab_two_rws_on_er_rg_proportion_avg = Simulator.simulate_dual_collab_on_er_rg_visited_prop(
        n, d, steps, walkers.RandomWalker, walkers.RandomWalker, k=k)
    single_rw_on_er_rg_prop_avg = Simulator.simulate_w_on_rg_visited_prop(n, d, 2*steps, walkers.RandomWalker, k=k)
    
    np.savetxt("collab_2_rws_on_er_rg_pa.csv", collab_two_rws_on_er_rg_proportion_avg, delimiter=',')
    np.savetxt("single_rw_on_er_rg_prop_avg.csv", single_rw_on_er_rg_prop_avg, delimiter=',')
```

```python
collab_two_rws_on_er_rg_proportion_avg = np.loadtxt("collab_2_rws_on_er_rg_pa.csv", delimiter=',')
single_rw_on_er_rg_prop_avg = np.loadtxt("single_rw_on_er_rg_prop_avg.csv", delimiter=',')
```

```python
t_max = len(collab_two_rws_on_er_rg_proportion_avg)

plt.plot(np.arange(1, t_max+1, step=1),
         [single_rw_on_er_rg_prop_avg[2*i] for i in range(t_max)], label='Single Walk: S(2t)')
plt.plot(np.arange(1, t_max+1, step=1),
         collab_two_rws_on_er_rg_proportion_avg, label='Collaboration - S1(t) + S2(t)')
plt.title("Proportion of Visited Vertices vs. t")
plt.xlabel("t")
plt.ylabel("Proportion")
plt.legend()
plt.show()
```

```python
diff = collab_two_rws_on_er_rg_proportion_avg - np.array([single_rw_on_er_rg_prop_avg[2*i] for i in range(t_max)])
print(f"Average Difference: {sum(diff)/len(diff)*100}%")
```

```python
plt.plot(np.arange(1, t_max+1, step=1), diff)
plt.title("Difference in Visited Vertices vs. t")
plt.xlabel("t")
plt.ylabel("Difference")
plt.show()
```

```python
steps = 3*n

if resimulate:
    collab_three_rws_on_er_rg_proportion_avg = Simulator.simulate_triple_collab_on_er_rg_visited_prop(
    n, d, steps, walkers.RandomWalker, walkers.RandomWalker, walkers.RandomWalker, k=k)
    single_rw_on_er_rg_prop_avg2 = Simulator.simulate_w_on_rg_visited_prop(n, d, 3*steps, walkers.RandomWalker, k=k)
    
    np.savetxt("collab_3_rws_on_er_rg_pa.csv", collab_three_rws_on_er_rg_proportion_avg, delimiter=',')
    np.savetxt("single_rw_on_er_rg_prop_avg2.csv", single_rw_on_er_rg_prop_avg2, delimiter=',')
```

```python
collab_three_rws_on_er_rg_proportion_avg = np.loadtxt("collab_3_rws_on_er_rg_pa.csv", delimiter=',')
single_rw_on_er_rg_prop_avg2 = np.loadtxt("single_rw_on_er_rg_prop_avg2.csv", delimiter=',')
```
