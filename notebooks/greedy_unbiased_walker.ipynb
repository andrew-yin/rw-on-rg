```python
# Setup
%matplotlib inline
import sys
import matplotlib.pyplot as plt
import numpy as np

sys.path.insert(0, '..')
plt.rcParams['figure.figsize'] = [10, 8]

from simulator import Simulator
```

```python
# Set parameters
n = 1000
expected_deg = 10
k = 100
p = expected_deg/n
```

```python
proportions = Simulator.simulate_guw_on_rg_visited_prop(n, p, k, comp_adj=True, d=expected_deg)
```

```python
plt.plot(np.arange(1, len(proportions)+1, step=1), proportions, label=f"n={n}, p={p}, k={k}")
plt.title("Proportion of Visited Vertices vs. # of Steps")
plt.xlabel("# of Steps")
plt.ylabel("Proportion of visited vertices")
plt.legend()
plt.show()
```
