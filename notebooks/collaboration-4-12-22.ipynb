```python
%matplotlib inline
import sys
import matplotlib.pyplot as plt
import numpy as np
sys.path.insert(0, '..')


plt.rcParams['figure.figsize'] = [10, 8]
from graph import Graph
from collaboration import *
```

```python
folder_path = "collaboration_4_12_22/"
simulate = True
```

## 2 RWs, ER Random Graph

```python
n = 100
d = 10
k = 100000
```

```python
if simulate:
    graph = Graph.get_er_random_graph(n, d/n)
    t = 5*n
    collab_covered_avg1, single_covered_avg1 = simulate_rw_collab_vs_single(
        graph=graph, 
        collab_num=2, 
        k=k, 
        collab_steps_per_iter=t)
    
    np.savetxt(folder_path+"2rw_vs_1_rw_on_er_collab.csv", collab_covered_avg1, delimiter=',')
    np.savetxt(folder_path+"2rw_vs_1_rw_on_er_single.csv", single_covered_avg1, delimiter=',')
else:
    collab_covered_avg1 = np.loadtxt(folder_path+"2rw_vs_1_rw_on_er_collab.csv", delimiter=',')
    single_covered_avg1 = np.loadtxt(folder_path+"2rw_vs_1_rw_on_er_single.csv", delimiter=',')
```

```python
c = 5
cs = np.arange(c, step=0.01)

scaled_single_covered_avg1 = np.array([single_covered_avg1[2*int(i*n)] for i in cs])
scaled_collab_covered_avg1 = np.array([collab_covered_avg1[int(i*n)] for i in cs])
plt.plot(cs, scaled_single_covered_avg1, label='Single Walk: S(2*c*n)')
plt.plot(cs, scaled_collab_covered_avg1, label='Collaboration: S1(c*n) + S2(c*n)')
plt.title("Visited Vertices vs. c")
plt.xlabel("c")
plt.ylabel("Visited Vertices")
plt.legend(title=f"n={n}, k={k}, ER")
plt.show()
```

```python
diff1 = collab_covered_avg1 - np.array([single_covered_avg1[2*i] for i in range(c*n+1)])
scaled_diff1 = np.array([diff1[int(i*n)] for i in cs])
plt.plot(cs, scaled_diff1, label='Difference: Collab - Single')
plt.title("Visited Vertices Difference vs. c")
plt.xlabel("c")
plt.ylabel("Visited Vertices")
plt.legend(title=f"n={n}, k={k}, ER")
plt.show()
```

```python
print("Is S1(t) + S2(t) >= S(2t) for all t? --", np.all(diff1 >= 0))
```
