## 2 RWs vs 1 RW on ER Random Graph

An ER random graph with $n = 1000$, $p = 10/1000$ is first generated—Then, we compare the vertices covered by two random walks over $t$ steps in collaboration vs. 1 random walk over $2t$ steps. This is run $k=10000$ times **on the same graph**, and the average is taken.

```python
n = 1000
d = 10
k = 10000
```

```python
if simulate:
    graph = Graph.get_er_random_graph(n, d/n)
    t = 5*n
    collab_covered_avg1, single_covered_avg1 = simulate_rw_collab_vs_single(
        graph=graph, 
        collab_num=2, 
        k=k, 
        collab_steps_per_iter=t)
    
    np.savetxt(folder_path+"2rw_vs_1_rw_on_er_collab.csv", collab_covered_avg1, delimiter=',')
    np.savetxt(folder_path+"2rw_vs_1_rw_on_er_single.csv", single_covered_avg1, delimiter=',')
else:
    collab_covered_avg1 = np.loadtxt(folder_path+"2rw_vs_1_rw_on_er_collab.csv", delimiter=',')
    single_covered_avg1 = np.loadtxt(folder_path+"2rw_vs_1_rw_on_er_single.csv", delimiter=',')
```

```python
t = 5*n

plt.plot(np.arange(t+1, step=1),
         [single_covered_avg1[2*i] for i in range(t+1)], label='Single Walk: S(2t)')
plt.plot(np.arange(t+1, step=1),
         collab_covered_avg1, label='Collaboration: S1(t) + S2(t)')
plt.title("Visited Vertices vs. t")
plt.xlabel("t")
plt.ylabel("Visited Vertices")
plt.legend(title=f"n={n}, k={k}, ER")
plt.show()
```

```python
diff1 = collab_covered_avg1 - np.array([single_covered_avg1[2*i] for i in range(t+1)])
plt.plot(np.arange(t+1, step=1), diff1, label='Difference: (S1(t) + S2(t)) - S(2t)')
plt.title("Visited Vertices vs. t")
plt.xlabel("t")
plt.ylabel("Visited Vertices")
plt.legend(title=f"n={n}, k={k}, ER")
plt.show()
```

```python
print("Is S1(t) + S2(t) >= S(2t) for all t? --", np.all(diff1 >= 0))
```

```python
print("What indices does this not hold?")
print(np.where(diff1 < 0))
```
