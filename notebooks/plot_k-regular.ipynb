```python
# Setup
%matplotlib inline
import sys
import matplotlib.pyplot as plt
import numpy as np

sys.path.insert(0, '..')
plt.rcParams['figure.figsize'] = [10, 8]

from graph import Graph
from walkers import RandomWalker
from simulator import Simulator
```

```python
n,k = 100, 1000
d = 10
p = 10/n
def plot_compare_er_kreg(n, k, d, p):
    steps = int(n*np.log(n)*np.log(n))
    a = Simulator.simulate_rw_on_rg_visited_prop(n, p, k, d=10)
    b = Simulator.simulate_rw_on_k_regular_visited_prop(d, n,steps, k)
    t = np.arange(1, len(b)+1, step=1)
    plt.plot(t, a, label='Erdos-Renyi')
    plt.plot(t, b, label='k-Regular')
    plt.title("ER vs. k-Regular: Proportion of Visited Vertices vs. # of Steps")
    plt.xlabel("# of Steps")
    plt.ylabel("Proportion of visited vertices")
    plt.hlines(1,0,len(t),colors='black',linestyles='dashed')
    plt.legend(title='n={:n}, '.format(n)+'k={:d}, '.format(k)+'steps={:d}'.format(steps))
    plt.show()
    
    
plot_compare_er_kreg(n,k,d,p)
```

```python
plt.xlim((100,550))
plt.ylim((0.6,1.1))
plot_compare_er_kreg(n,k,d,p)
```

```python
n,k = 100, 1000
d =10
p = 10/n
def plot_compare_strats_kreg(n, k, d, p):
    steps = int(n*np.log(n)*np.log(n))
    kreg_rand = Simulator.simulate_rw_on_k_regular_visited_prop(d, n,steps, k)
    kreg_greed = Simulator.simulate_guw_on_k_regular_visited_prop(d,n,steps, k)
    t = np.arange(1, len(kreg_rand)+1, step=1)
    plt.plot(t, kreg_rand, label='k-Regular: random')
    plt.plot(t, kreg_greed, label='k-Regular: greedy')
    plt.title("strategies on k-Regular: Proportion of Visited Vertices vs. # of Steps")
    plt.xlabel("# of Steps")
    plt.ylabel("Proportion of visited vertices")
    plt.hlines(1,0,len(t),colors='black',linestyles='dashed')
    plt.legend(title='n={:n}, '.format(n)+'k={:d}, '.format(k)+'steps={:d}'.format(steps))
    plt.show()
    
plot_compare_strats_kreg(n,k,d,p)
```

```python
plt.xlim((0,550))
plt.ylim((0.4,1.1))
plot_compare_strats_kreg(n,k,d,p)
```
